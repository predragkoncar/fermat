**Fermat's Last Theorem**

There are no three positive integers $x, y, z$ that satisfy the equation $x^n+y^n=z^n$ for any integer value of $n>2$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ints = np.arange(1,256)

z = np.zeros((len(ints),len(ints)))

# run the simulation
for x in ints:
    for y in ints:
        z[x-1,y-1] = np.cbrt((x**3 + y**3))
```

```python
zInts = z%1 == 0
```

```python
# visualize
fig,ax = plt.subplots(1,2,figsize=(12,6))
fig.tight_layout(pad=5)

# resulting z-values
h = ax[0].imshow(z)
fig.colorbar(h,ax=ax[0],fraction=.045)
ax[0].set_xlabel('y')
ax[0].set_ylabel('x')
ax[0].set_title('$z=\sqrt[3]{x^3+y^3}$',fontsize=16)

# boolean integer map
h = ax[1].imshow(zInts,vmin=0,vmax=1)
fig.colorbar(h,ax=ax[1],fraction=.045)
ax[1].set_xlabel('y')
ax[1].set_ylabel('x')
ax[1].set_title('Whether $z$ is integer',fontsize=16)

plt.show()
```
